## Univariate Analysis

```python
df_Tilapia=df.loc[df['Species']=='Tilapia']
```

```python
df_Nilotica=df.loc[df['Species']=='Nilotica']
df_Taki=df.loc[df['Species']=='Taki']
df_Magur=df.loc[df['Species']=='Magur']
df_Baim=df.loc[df['Species']=='Baim']
df_Rui=df.loc[df['Species']=='Rui']
```

```python
plt.plot(df_Tilapia['Length'],np.zeros_like(df_Tilapia['Length']),'o')
plt.plot(df_Nilotica['Length'],np.zeros_like(df_Nilotica['Length']),'o')
plt.plot(df_Taki['Length'],np.zeros_like(df_Taki['Length']),'o')
plt.plot(df_Magur['Length'],np.zeros_like(df_Magur['Length']),'o')
plt.plot(df_Baim['Length'],np.zeros_like(df_Baim['Length']),'o')
plt.plot(df_Rui['Length'],np.zeros_like(df_Rui['Length']),'o')
plt.xlabel('Fish length')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue="Species").map(plt.scatter,"Length", "Height").add_legend();
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="Species",size=3)
```
