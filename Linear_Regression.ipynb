```python
df = pd.read_csv('/content/Fisheries_DataAnalysis_Projects/Clam/clam.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
X = df[['SL', 'SH', 'SW']].copy()
y = df[['Total.wt']].copy()
```

```python
X
```

```python
y
```

```python
X["intercept"] = 1  # set intercept value as 1
```

```python
X = X[["intercept", 'SL', 'SH', 'SW']]   # reorder X column
```

```python
X
```

```python
X_T = X.T  # transposing X
```

```python
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ y  # Calculation of coefficient
```

```python
B
```

```python
B.index = X.columns  # Replace 0 1 2 into Column name
```

```python
B
```

```python
predictions = X @ B
```

```python
predictions
```

```python
# sum of square Residues; it'll tell how big our error are, from the best fit line
# measuring variance
SSR = ((y - predictions) ** 2).sum()
```

```python
SSR
```

```python

# sum of square Total; difference btw y value and mean y values
TSS = ((y - y.mean()) ** 2).sum()
```

```python
TSS
```

```python
# predecting accuracy of model
R2 = 1 - (SSR / TSS)
R2
```
